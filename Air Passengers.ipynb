```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Portfolio\hands on machine learning dataset\airpassenger.xlsx")
```

```python
df['step'] = range(len(df))
# 쌓으면서 더하기
df['cum_pass'] = df['#Passengers'].cumsum()
# 쌓은것중 가장 큰값
df['cum_max'] = df['#Passengers'].cummax()
# 쌓은것중 가장 작은값
df['cum_min'] = df['#Passengers'].cummin()
```

```python
df.head(5)
```

## 년월일 분석

```python
df['Month']= df['Month'].astype('str')
```

```python
temp_date = df['Month'].map(lambda x : x.split("-"))
temp_date = np.array(temp_date.values.tolist())
temp_date[:5]
```

```python
df['year'] = temp_date[:,0]
df['month'] = temp_date[:,1]
df['day'] = temp_date[:,2]
```

```python
df.head(5)
```

```python
# #Passenger의 차이
df['diff'] = df['#Passengers'].diff().fillna(0)
```

```python
df.head()
```

```python
df['pct']= df['#Passengers'].pct_change().map(lambda x : x*100).round(2)
```
