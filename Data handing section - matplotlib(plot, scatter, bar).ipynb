```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = range(100) # 0부터 100까지
y = [value ** 2 for value in x] # x 제곱
```

```python
# 플롯 그리기
plt.plot(x,y)
plt.show()
```

### matplotlib 특징

- figure 객체에 생성<br>
- pyplot 객체 사용시, 기본 figure에 그래프가 생성

```python
# 기본적인 그래프
x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

x_2 = range(100)
y_2 = [np.sin(value) for value in x_2]

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.show()
```

```python
# figure 사용한 subplot
fig = plt.figure() # 새로운 캠퍼스 생성
fig.set_size_inches(10,5) # 크기 지정
# subplot 생성
ax_1 = fig.add_subplot(1,2,1)  # 가로 1, 세로 2인 1번째 그래프
ax_2 = fig.add_subplot(1,2,2)  # 거로 1 ,세로 2인 2번째 그래프

# 플롯 그리기
ax_1.plot(x_1,y_1, c = 'b') #c = 색설정 'b' -> blue
ax_2.plot(x_2,y_2, c = 'g', ls = '--') #c = 색설정 'g' -> green   #ls(linestyle) 
plt.show()
```

### 범례

```python
# x,y 범위 설정
x = range(100)
y = range(100,200)
x_1 = range(100)
y_1 = range(0,100)

# 
plt.plot(x,y, c = "g", label = "line 1")
plt.plot(x_1,y_1, c = "r", label = "line 2")
plt.legend(loc = "lower right")
plt.title("legend graph")
plt.xlabel("x_graph")
plt.ylabel("y_graph")
plt.show()
```
